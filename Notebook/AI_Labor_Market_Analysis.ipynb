```python
def clean_domain(val):
    val = str(val).strip()
    if val == 'nan' or val == '':
        return 'Unknown'
    if ',' in val:
        return 'Multimodal'
    return val


df_final['Domain'] = df_final['Domain'].apply(clean_domain)

display(df_final)
```

```python
columns = ['Publication date', 'Compute_Log10', 'Parameters',
           'Domain', 'Organization', 'Country (of organization)']

df_final = df_final[columns]

df_final
```

```python
print(df_final['Domain'].unique())
```

```python
domain_mapping = {
    'Language': 'Language & Audio',
    'Speech': 'Language & Audio',
    'Audio': 'Language & Audio',
    'Search': 'Language & Audio',
    'Recommendation': 'Language & Audio',
    'Vision': 'Vision & Media',
    'Video': 'Vision & Media',
    'Image generation': 'Vision & Media',
    'Multimodal': 'Multimodal',
    'Biology': 'Science & Engineering',
    'Earth science': 'Science & Engineering',
    'Materials science': 'Science & Engineering',
    'Medicine': 'Science & Engineering',
    'Mathematics': 'Science & Engineering',
    'Robotics': 'Robotics & Games',
    'Driving': 'Robotics & Games',
    'Games': 'Robotics & Games',
    'Other': 'Other',
    'Unknown': 'Other'
}


df_final['Macro_Domain'] = df_final['Domain'].map(domain_mapping)

columns = ['Publication date', 'Compute_Log10', 'Parameters',
           'Macro_Domain', 'Organization', 'Country (of organization)']

df_final = df_final[columns]

df_final = df_final.rename(columns={
    'Macro_Domain' : 'Domain'
})

display(df_final)
```

```python
print(df_final['Country (of organization)'].unique())
```

```python
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Czechia', 'Denmark', 
    'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 
    'Ireland', 'Italy', 'Lithuania', 'Netherlands', 'Poland', 'Portugal', 
    'Russia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine', 
    'United Kingdom of Great Britain and Northern Ireland'
]

def map_region(country_str):
    if pd.isna(country_str):
        return None
    
    # Prende il primo paese della lista
    first_country = country_str.split(',')[0].strip()
    
    if first_country == 'United States of America':
        return 'United States'
    elif first_country == 'China':
        return 'China'
    elif first_country in european_countries:
        return 'Europe'
    else:
        return None


df_final['Region'] = df_final['Country (of organization)'].apply(map_region)


df_final = df_final.dropna(subset=['Region']).copy()
df_final = df_final.drop(columns=['Country (of organization)'])

display(df_final)
```

```python
# ai_jobs_trends_dataset file
file_path = os.path.join(LABOR_FOLDER, 'ai_job_trends_dataset.csv')
df_ai = pd.read_csv(file_path)

print('--- DF Head ---')
display(df_ai.head())

print('\n --- DF Info')
df_ai.info()
```

```python
print(df_ai['Location'].unique())
```

```python
columns = ['Job Title', 'Industry', 'Job Status', 
           'AI Impact Level', 'Required Education',
           'Job Openings (2024)', 'Projected Openings (2030)',
           'Automation Risk (%)', 'Location']

df_filtered = df_ai[columns].copy()


europe_countries = ['UK', 'Germany']
target_countries = europe_countries + ['China', 'USA']
df_filtered = df_filtered[df_filtered['Location'].isin(target_countries)]

df_filtered['Location'] = df_filtered['Location'].apply(
    lambda x: "Europe" if x in europe_countries else x
)

df_filtered['Location'] = df_filtered['Location'].replace('USA', 'United States')
df_filtered = df_filtered.rename(columns={
        'Location' : 'Region'
})

df_filtered
```

```python
agg_rules = {
    'Job Openings (2024)' : 'sum',
    'Projected Openings (2030)' : 'sum',
    'Automation Risk (%)' : 'mean',
    'AI Impact Level' : 'first',
    'Job Status' : 'first',
    'Required Education' : 'first'
}

df_final = df_filtered.groupby(['Region', 'Industry', 'Job Title']).agg(agg_rules).reset_index()

df_finale = df_final.round(2)

df_finale
```

```python
df_finale['Job Growth'] = df_finale['Projected Openings (2030)'] - df_finale['Job Openings (2024)']

df_finale
```

```python
# last excel files
excel_path = os.path.join(LABOR_FOLDER, 'AIOE_DataAppendix.xlsx')
output_path = os.path.join(CLEAN_FOLDER, 'AIOE_Data_Cleaned.xlsx')


def get_sector(naics, title):
    naics = str(naics)
    title = str(title).lower()
    if naics.startswith(('334', '333', '335')): return 'Infrastructure (Chip)'
    if naics.startswith(('517', '518', '519')): return 'Cloud & Ecosystems'
    if naics.startswith('511') or 'software' in title: return 'Software & Platform'
    if naics.startswith('5415') or 'data' in title: return 'Data & Storage'
    if naics.startswith(('52', '53', '54', '55', '56', '61', '62', '92')): return 'Business Services'
    if naics.startswith(('42', '44', '45', '71', '72', '81')): return 'Consumer & Retail'
    return 'Infrastructure (General)'


with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
    
    # --- APPENDIX A ---
    df_a = pd.read_excel(excel_path, sheet_name='Appendix A')

    df_a['AIOE'] = pd.to_numeric(df_a['AIOE'], errors='coerce')
    df_a['AIOE'] = df_a['AIOE'].apply(lambda x: round(x/1000000, 3) if abs(x) > 100 else round(x, 3))
    
    soc_map = {'11':'Management','13':'Business','15':'Computer/Math','17':'Eng','19':'Science','23':'Legal','25':'Education','27':'Arts','29':'Health','31':'Health Supp','41':'Sales','43':'Admin'}
    df_a['Macro-Category'] = df_a['SOC Code'].str[:2].map(soc_map).fillna('Other Services')
    df_a.to_excel(writer, sheet_name='Appendix_A_Cleaned', index=False)

    # --- APPENDIX B ---
    df_b = pd.read_excel(excel_path, sheet_name='Appendix B')
    df_b['AIIE'] = pd.to_numeric(df_b['AIIE'], errors='coerce')
    df_b['AIIE'] = df_b['AIIE'].apply(lambda x: round(x/1000000, 3) if abs(x) > 100 else round(x, 3))
    df_b['Sector'] = df_b.apply(lambda x: get_sector(x['NAICS'], x['Industry Title']), axis=1)
    df_b.to_excel(writer, sheet_name='Appendix_B_Cleaned', index=False)

    # --- APPENDIX E ---
    df_e = pd.read_excel(excel_path, sheet_name='Appendix E')
    col_e = 'Ability-Level AI Exposure'
    df_e[col_e] = pd.to_numeric(df_e[col_e], errors='coerce')
    df_e[col_e] = df_e[col_e].apply(lambda x: round(x/1000000, 3) if abs(x) > 100 else round(x, 3))
    
    def get_ab_cat(ab):
        if any(w in ab for w in ['Strength', 'Flexibility', 'Dexterity']): return 'Physical'
        if any(w in ab for w in ['Vision', 'Hearing', 'Speech']): return 'Sensory'
        return 'Cognitive'
    
    df_e['Macro-Category'] = df_e['O*NET Abilities'].apply(get_ab_cat)
    df_e.to_excel(writer, sheet_name='Appendix_E_Cleaned', index=False)

print(f"✅ FINALMENTE! File Excel creato: {output_path}")
```

```python
# Nome esatto del file Excel
FILE_NAME = 'AIOE_Data_Cleaned.xlsx'
FILE_PATH = os.path.join(LABOR_FOLDER, FILE_NAME)

def elabora_e_salva_excel():
    if not os.path.exists(FILE_PATH):
        print(f"Errore: Il file {FILE_NAME} non è stato trovato in {LABOR_FOLDER}")
        return

   
    try:
        # Aggiornato con il nome trovato: Appendix_A_Cleaned
        df = pd.read_excel(FILE_PATH, sheet_name='Appendix_A_Cleaned')
        print(f"✅ Foglio 'Appendix_A_Cleaned' caricato con successo.")
    except Exception as e:
        print(f"Errore durante l'apertura: {e}")
        return

    df.columns = df.columns.str.strip()
    df['Macro-Category'] = df['Macro-Category'].astype(str).str.strip()

    mappatura_cluster = {
        # STEM & Analysis
        'Computer/Math': 'STEM & Analysis',
        'Eng': 'STEM & Analysis',
        'Science': 'STEM & Analysis',
        
        # Business & Legal
        'Management': 'Business & Legal',
        'Business': 'Business & Legal',
        'Legal': 'Business & Legal',
        'Sales': 'Business & Legal',
        
        # Education & Care
        'Education': 'Education & Care',
        'Health': 'Education & Care',
        'Health Supp': 'Education & Care',
        'Arts': 'Education & Care',
        
        # Operations & Services
        'Admin': 'Operations & Services',
        'Other Services': 'Operations & Services'
    }


    df['Broad_Category'] = df['Macro-Category'].map(mappatura_cluster).fillna('Other')


    output_path = os.path.join(CLEAN_FOLDER, 'Appendix_A_Categorized.csv')
    df.to_csv(output_path, index=False, encoding='utf-8-sig')
    
    print(f"✅ File salvato in: {output_path}")
    

    print("\nConteggio righe per Nuova Categoria:")
    print(df['Broad_Category'].value_counts())

if __name__ == "__main__":
    elabora_e_salva_excel()
```

```python
FILE_NAME = 'AIOE_Data_Cleaned.xlsx'
FILE_PATH = os.path.join(LABOR_FOLDER, FILE_NAME)

def esporta_appendici_excel():
    if not os.path.exists(FILE_PATH):
        print(f"Errore: File {FILE_NAME} non trovato in {LABOR_FOLDER}")
        return

    # Lista dei fogli da estrarre
    fogli_da_estrarre = ['Appendix_A_Cleaned', 'Appendix_B_Cleaned', 'Appendix_E_Cleaned']
    
   
    mappatura_cluster = {
        'Computer/Math': 'STEM & Analysis', 'Eng': 'STEM & Analysis', 'Science': 'STEM & Analysis',
        'Management': 'Business & Legal', 'Business': 'Business & Legal', 'Legal': 'Business & Legal', 'Sales': 'Business & Legal',
        'Education': 'Education & Care', 'Health': 'Education & Care', 'Health Supp': 'Education & Care', 'Arts': 'Education & Care',
        'Admin': 'Operations & Services', 'Other Services': 'Operations & Services'
    }

    for foglio in fogli_da_estrarre:
        try:
            print(f"Esportazione foglio: {foglio}...")
            
            df = pd.read_excel(FILE_PATH, sheet_name=foglio)
            
         
            df.columns = df.columns.str.strip()
            
            
            if foglio == 'Appendix_A_Cleaned':
                if 'Macro-Category' in df.columns:
                    df['Macro-Category'] = df['Macro-Category'].astype(str).str.strip()
                    df['Broad_Category'] = df['Macro-Category'].map(mappatura_cluster).fillna('Other')
                else:
                    print(f"Attenzione: Colonna 'Macro-Category' non trovata in {foglio}")

            
            output_name = f"{foglio}.xlsx"
            output_path = os.path.join(CLEAN_FOLDER, output_name)
            
            
            df.to_excel(output_path, index=False, engine='openpyxl')
            print(f"✅ File Excel creato: {output_path}")

        except Exception as e:
            print(f"❌ Errore durante l'esportazione di {foglio}: {e}")

if __name__ == "__main__":
    esporta_excel_separati = esporta_appendici_excel()
```
